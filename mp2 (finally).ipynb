```python
baselineT = baseline_csv.transpose()
metforminT = metformin_csv.transpose()
geneNames = set(baseline_csv['GeneID'].tolist()).intersection(set(metformin_csv['GeneID'].tolist()))
print("%d genes common in both dataset" % len(geneNames))
numDifferentlyExpressedGene = 0
candidateGenes = set()
p_values=[]
for selectGene in geneNames:
    bIndx = baselineT[1:2:].transpose()['GeneID'].tolist().index(selectGene)
    mIndx = metforminT[1:2:].transpose()['GeneID'].tolist().index(selectGene)
    b = np.log(baselineT[bIndx].iloc[6:169].astype(float))
    m = np.log(metforminT[mIndx].iloc[6:169].astype(float))
    Vars = {
        "Baseline": b,
        "Metformin": m
    }
    testResult = stats.ks_2samp(Vars['Baseline'], Vars["Metformin"])
    if testResult[1] < 1e-1:
        candidateGenes.add(selectGene)
        numDifferentlyExpressedGene += 1
        p_values.append(testResult[1])
#        print("%d: %s gene is differently expressed in baseline and metformin treated cells" % (numDifferentlyExpressedGene, selectGene))
#        print("Test Statistics %f, %g" % (testResult[0], testResult[1]))
#0.10,0.05,0.025,0.01,0.005  and 0.001
length=len(p_values)
alpha_01=0
for i in range(0,length):
    if(p_values[i]>0.1):
        alpha_01+=1
alpha_005=0
for i in range(0,length):
    if(p_values[i]>0.05):
        alpha_005+=1
alpha_0025=0
for i in range(0,length):
    if(p_values[i]>0.025):
        alpha_0025+=1
alpha_001=0
for i in range(0,length):
    if(p_values[i]>0.01):
        alpha_001+=1
alpha_0005=0
for i in range(0,length):
    if(p_values[i]>0.005):
        alpha_0005+=1
alpha_0001=0
for i in range(0,length):
    if(p_values[i]>0.001):
        alpha_0001+=1
fail01=length-alpha_01
fail005=length-alpha_005
fail0025=length-alpha_0025
fail001=length-alpha_001
fail0005=length-alpha_0005
fail0001=length-alpha_0001
#0.10,0.05,0.025,0.01,0.005  and 0.001
```

```python
bt_mean_vec = np.mean(bt_std, axis=0) # by column
cov_bt = (bt_std - bt_mean_vec).T.dot((bt_std - bt_mean_vec)) / (bt_std.shape[0] - 1)
print(cov_bt)
```

```python
bt_mat = np.cov(bt_std.T)

bt_eig_vals, bt_eig_vecs = np.linalg.eig(cov_bt)
print(bt_std)
```

```python
print(bt_eig_vals)
```

```python
print(bt_eig_vecs.T)
```

```python
bt_eig_pairs = [(np.abs(bt_eig_vals[i]), bt_eig_vecs[:,i]) for i in range(len(bt_eig_vals))]
bt_eig_pairs.sort(key=lambda x: x[0], reverse=True)
```

```python
print('Eigenvalues in descending order:')
for i in bt_eig_pairs:
    print(i[0],": ",i[1])
```

```python
bt_total = sum(bt_eig_vals)
bt_var_exp = [(i / bt_total)*100 for i in sorted(bt_eig_vals, reverse=True)]
bt_cum_var_exp = np.cumsum(bt_var_exp)
bt_cum_var_exp
```

```python
with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(6, 4))

    plt.bar(range(1170), bt_var_exp, alpha=0.5, align='center',
            label='individual explained variance')
    plt.step(range(1170), bt_cum_var_exp, where='mid',
             label='cumulative explained variance')
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal components')
    plt.legend(loc='best')
    plt.tight_layout()
    plt.show()
```
